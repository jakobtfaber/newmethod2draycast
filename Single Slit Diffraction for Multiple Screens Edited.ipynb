```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random
from scipy.stats import kde
import pandas as pd
import seaborn as sns


random.seed(1234234345)

nrays = 100000

x = np.zeros(nrays)
y = np.zeros(nrays)
x1 = np.zeros(nrays)
y1 = np.zeros(nrays)
x2 = np.zeros(nrays)
y2 = np.zeros(nrays)
x3 = np.zeros(nrays)
y3 = np.zeros(nrays)
x4 = np.zeros(nrays)
y4 = np.zeros(nrays)

slit_width = 0.01

a = -(slit_width / 2)
b = (slit_width / 2)

for i in range(nrays):
    
    r = 1

    theta = random.gauss(0, np.pi/2)
    phi = random.gauss(0, np.pi/2)
    theta2 = random.gauss(0, np.pi/2)
    phi2 = random.gauss(0, np.pi/2)
    theta3 = random.gauss(0, np.pi/2)
    phi3 = random.gauss(0, np.pi/2)
    theta4 = random.gauss(0, np.pi/2)
    phi4 = random.gauss(0, np.pi/2)
    
    x[i] = r * np.cos(theta) * np.sin(phi)
    y[i] = r * np.sin(theta) * np.sin(phi)
    
    if a <= x[i] <= b:

        x1[i] = r * np.cos(theta) * np.sin(phi)
        y1[i] = r * np.sin(theta) * np.sin(phi)
    
        if x[i] != 0:
        
            x2[i] = ((r) * np.cos(theta2) * np.sin(phi2)) + (x[i])
            
        if y[i] != 0:
            
            y2[i] = ((r) * np.sin(theta2) * np.sin(phi2)) + (y[i])
            
    if x2[i] != 0:
        
        x3[i] = ((r) * np.cos(theta3) * np.sin(phi3)) + (x2[i])
        
    if y2[i] != 0:
        
        y3[i] = ((r) * np.sin(theta3) * np.sin(phi3)) + (y2[i])
        
    if x3[i] != 0:
        
        x4[i] = ((r) * np.cos(theta4) * np.sin(phi4)) + (x3[i])
        
    if y3[i] != 0:
        
        y4[i] = ((r) * np.sin(theta4) * np.sin(phi4)) + (y3[i])    
            
fig = plt.figure(figsize = (24,16))

ax1 = fig.add_subplot(3,5,1)
ax1.set_title(str(nrays) + " rays: Beam (Pre-Screen)")
ax1.set_ylabel("Distance")
ax1.set_xlabel("Distance")
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.scatter(x,y, c = 'black', marker = 'o')
plt.gca().set_aspect('equal')

ax2 = fig.add_subplot(3,5,2)
ax2.set_title(str(nrays) + " rays: Slit in Screen 1")
ax2.set_ylabel("Distance")
ax2.set_xlabel("Distance")
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.scatter(x1,y1, c = 'black', marker = 'o')
plt.gca().set_aspect('equal')

ax3 = fig.add_subplot(3,5,3)
ax3.set_title(str(nrays) + " rays: Diffraction Pattern on Screen 2")
ax3.set_ylabel("Distance")
ax3.set_xlabel("Distance")
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.scatter(x2,y2, c = 'black', marker = 'o')
plt.gca().set_aspect('equal')

ax4 = fig.add_subplot(3,5,4)
ax4.set_title(str(nrays) + " rays: Diffraction Pattern on Screen 3")
ax4.set_ylabel("Distance")
ax4.set_xlabel("Distance")
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.scatter(x3,y3, c = 'black', marker = 'o')
plt.gca().set_aspect('equal')

ax4 = fig.add_subplot(3,5,5)
ax4.set_title(str(nrays) + " rays: Diffraction Pattern on Screen 3")
ax4.set_ylabel("Distance")
ax4.set_xlabel("Distance")
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.scatter(x4,y4, c = 'black', marker = 'o')
plt.gca().set_aspect('equal')

ax5 = fig.add_subplot(3,5,6, aspect = 1.85)
ax5.set_title(str(nrays) + " rays: X Array")
ax5.set_ylabel("Point Density")
ax5.set_xlabel("Distance")
sns.kdeplot(x, shade=True, color = "r")

ax6 = fig.add_subplot(3,5,7, aspect = 1.85)
ax6.set_title(str(nrays) + " rays: Y Array")
ax6.set_ylabel("Point Density")
ax6.set_xlabel("Distance")
sns.kdeplot(y, shade=True, color = "r")

ax7 = fig.add_subplot(3,5,8, aspect = 0.0000001)
ax7.set_title(str(nrays) + " rays: X1 Array")
ax7.set_ylabel("Point Density")
ax7.set_xlabel("Distance")
plt.xlim(-0.001,0.001)
sns.kdeplot(x1, shade=True, color = "r")

ax8 = fig.add_subplot(3,5,9, aspect = 0.05)
ax8.set_title(str(nrays) + " rays: Y1 Array")
ax8.set_ylabel("Point Density")
ax8.set_xlabel("Distance")
sns.kdeplot(y1, shade=True, color = "r")

ax9 = fig.add_subplot(3,5,10, aspect = 0.03)
ax9.set_title(str(nrays) + " rays: X2 Array")
ax9.set_ylabel("Point Density")
ax9.set_xlabel("Distance")
sns.kdeplot(x2, shade=True, color = "r")

ax10 = fig.add_subplot(3,5,11, aspect = 0.065)
ax10.set_title(str(nrays) + " rays: Y2 Array")
ax10.set_ylabel("Point Density")
ax10.set_xlabel("Distance")
sns.kdeplot(y2, shade=True, color = "r")

ax11 = fig.add_subplot(3,5,12, aspect = 0.065)
ax11.set_title(str(nrays) + " rays: X3 Array")
ax11.set_ylabel("Point Density")
ax11.set_xlabel("Distance")
sns.kdeplot(x3, shade=True, color = "r")

ax12 = fig.add_subplot(3,5,13, aspect = 0.18)
ax12.set_title(str(nrays) + " rays: Y3 Array")
ax12.set_ylabel("Point Density")
ax12.set_xlabel("Distance")
sns.kdeplot(y3, shade=True, color = "r")

ax12 = fig.add_subplot(3,5,14, aspect = 0.18)
ax12.set_title(str(nrays) + " rays: X4 Array")
ax12.set_ylabel("Point Density")
ax12.set_xlabel("Distance")
sns.kdeplot(x4, shade=True, color = "r")

ax12 = fig.add_subplot(3,5,15, aspect = 0.3)
ax12.set_title(str(nrays) + " rays: Y4 Array")
ax12.set_ylabel("Point Density")
ax12.set_xlabel("Distance")
sns.kdeplot(y4, shade=True, color = "r")

plt.tight_layout()
plt.show()
```
